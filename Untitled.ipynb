## Разбиваем данные на обучающую и тренировочную выборки

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=21)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 10)
knn.fit(X_train, y_train)
```

```python
y_train_predict = knn.predict(X_train)
y_test_predict = knn.predict(X_test)

err_train = np.mean(y_train != y_train_predict)
err_test  = np.mean(y_test  != y_test_predict)

print(err_train, err_test)
```

```python
#Из экспериментов с различным числом ближайших соседей можно сделать вывод, что n_neighbors = 10 - оптимальное для нас значение.
```

# AdaBoost

```python
from sklearn import ensemble
ada = ensemble.AdaBoostClassifier(n_estimators = 50)
ada.fit(X_train, y_train)

err_train = np.mean(y_train != ada.predict(X_train))
err_test = np.mean(y_test != ada.predict(X_test))

print(err_train, err_test)
```

## Подбор параметров для kNN 

```python
from sklearn.model_selection import GridSearchCV

nnb = [1, 3, 5, 10, 15, 20, 25, 35, 45, 55]
knn = KNeighborsClassifier()
grid = GridSearchCV(knn, param_grid = {'n_neighbors': nnb}, cv=10)
grid.fit(X_train, y_train)

best_cv_err = 1 - grid.best_score_
best_n_neighbors = grid.best_estimator_.n_neighbors
print(best_cv_err, best_n_neighbors)
```

```python
knn = KNeighborsClassifier(n_neighbors = best_n_neighbors).fit(X_train, y_train)

err_train = np.mean(y_train != knn.predict(X_train))
err_test  = np.mean(y_test  != knn.predict(X_test))

print(err_train, err_test)
```

#### При n_neighbors = 1 происходит переобучение, следовательно ошибка для x_train будет равняться 0, а для тестовой ошибка будет гораздо больше. Поэтому выберем число соседей равное 10, как мы уже получили ранее. 

```python
ran_k = range(1, 15)
err_train_mas = []
err_test_mas = []
for k in ran_k:
    mod_k = KNeighborsClassifier(n_neighbors = k)
    mod_k.fit(X_train, y_train)
    err_train_mas.append(np.mean(mod_k.predict(X_train) != y_train))
    err_test_mas.append(np.mean(mod_k.predict(X_test) != y_test))
plt.plot(ran_k, err_train_mas, 'r-o', label = 'Train error')
plt.plot(ran_k, err_test_mas,  'b-o', label = 'Test error')    
plt.legend(loc = 2)
```

```python
#На графике видно, что ошибка на тестовой и на тренировочной выборке при n = 10 почти совпадают, а далее
#при увелечении числа соседей, ошибка на тренировочной выборке растет, а на тестовой падает.
```

### PCA 

```python
from sklearn.decomposition import PCA


components_arr = np.arange(1, 15)
test_err = []
train_err = []
oob_err = []

for components in components_arr:
    pca = PCA(n_components=components)
    X_pca = pca.fit_transform(X)
    X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.3, random_state=21)
    knn = KNeighborsClassifier(n_neighbors=7)
    knn.fit(X_train, y_train)
    
    train_err.append(np.mean(y_train != knn.predict(X_train)))
    test_err.append(np.mean(y_test  != knn.predict(X_test)))
    

plt.figure()
plt.plot(components_arr, train_err, 'b-o', label = 'train')
plt.plot(components_arr, test_err, 'r-o', label = 'test')
plt.xlim([np.min(components_arr), np.max(components_arr)])
plt.title('PCA')
plt.xlabel('components')
plt.ylabel('error')
plt.legend()
plt.show()
```

```python
#Применение метода главных компонент позволило снизить ошибку при предсказании вектора y.
```

### Подбор гиперпараметров

подбор alpha

```python
alpha_arr = np.logspace(-3, 2, 21)
test_err = []
train_err = []
for alpha in alpha_arr:
    mlp_model = MLPClassifier(alpha = alpha, hidden_layer_sizes = (100,), 
                              solver = 'lbfgs', activation = 'relu', random_state = 27)
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.semilogx(alpha_arr, train_err, 'b-o', label = 'test')
plt.semilogx(alpha_arr, test_err, 'r-o', label = 'train')
plt.xlim([np.max(alpha_arr), np.min(alpha_arr)])
plt.title('Error vs. alpha')
plt.xlabel('alpha')
plt.ylabel('error')
plt.legend()
pass
```

При очень малых альфа до ~0.5 происходит переобучение, т.к. на на графике видно, что ошибка для обучающей выборки составляет 0.

При альфа > 10 наблюдаем уже недообучение.

Минимальная ошибка

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)
print(min_train_err, min_test_err)
```

Подбираем оптимальное количество нейронов

```python
nodes_arr = np.arange(5, 100, 5)
test_err = []
train_err = []
for nodes in nodes_arr:
    mlp_model = MLPClassifier(hidden_layer_sizes = (nodes,), solver = 'lbfgs',
                              activation = 'relu', random_state = 27)
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
```

```python
plt.semilogx(nodes_arr, train_err, 'b-o', label = 'train')
plt.semilogx(nodes_arr, test_err, 'r-o', label = 'test')
plt.xlim([np.max(nodes_arr), np.min(nodes_arr)])
plt.title('Error vs. nodes count')
plt.xlabel('nodes')
plt.ylabel('error')
plt.legend()
```

Минимально достигнутая ошибка

```python
print(np.min(train_err), np.min(test_err))
```

Оптимальное количество нейронов

```python
print(nodes_arr[test_err == min_test_err][0])
```

Наилучшее значение ошибки для тестовой выборки было достигнуто при 90 нейронов и составило ~0.008. При этом наблюдается переобучение, т.к. ошибка на тренировочной выборке при любых значениях числа нейронов составила 0.

# Random forest

```python
from sklearn import ensemble
rf = ensemble.RandomForestClassifier(n_estimators = 100)
rf.fit(X_train, y_train)
```

```python
# Считаем среднюю ошибку
err_train = np.mean(y_train != rf.predict(X_train))
err_test  = np.mean(y_test  != rf.predict(X_test))
print(err_train, err_test)
```

```python

trees_arr = np.arange(25, 126, 5)
test_err = []
train_err = []
oob_err = []
for trees in trees_arr:
    rf = ensemble.RandomForestClassifier(n_estimators = trees, oob_score=True)
    rf.fit(X_train, y_train)
    train_err.append(np.mean(y_train != rf.predict(X_train)))
    test_err.append(np.mean(y_test  != rf.predict(X_test)))
    oob_err.append(1 - rf.oob_score_)
```

```python
plt.plot(trees_arr, train_err, 'b-o', label = 'train')
plt.plot(trees_arr, test_err, 'r-o', label = 'test')
plt.plot(trees_arr, oob_err, 'g-o', label = 'oob')
plt.xlim([np.max(trees_arr), np.min(trees_arr)])
plt.title('Error vs. number of trees')
plt.xlabel('trees')
plt.ylabel('error')
plt.legend()
```

# Gradient Boosted Trees

```python
err_train = np.mean(y_train != gbt.predict(X_train))
err_test  = np.mean(y_test  != gbt.predict(X_test))
print(err_train, err_test)
```

```python
plt.plot(trees_arr, train_err, 'b-o', label = 'train')
plt.plot(trees_arr, test_err, 'r-o', label = 'test')
plt.xlim([np.max(trees_arr), np.min(trees_arr)])
plt.title('Error vs. number of trees')
plt.xlabel('trees')
plt.ylabel('error')
plt.legend()
```

Наблюдается сильное переобучение, т.к. для обучаещей выборки ошибка = 0
